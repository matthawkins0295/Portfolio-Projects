```python
# Correlation Matrix between all numeric columns 
df1.corr(method = 'pearson',numeric_only = True)
```

```python
correlation_matrix = df1.corr(method = 'pearson',numeric_only = True)
```

```python
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Figures')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Assinging numeric values to non-numeric columns
df_numerized = df1.copy() 

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.apply(lambda x: x.factorize()[0]).corr(method='pearson')
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")

plt.title('Correlation Matrix for Numeric Figures')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()
```

```python
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#looking for columns with high correlation
high_corr = sorted_pairs[(sorted_pairs) >.5]
high_corr
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df1.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted['gross'] = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# Group by 'company' and 'year' and sum 'gross'
CompanyGrossSumYear = df1.groupby(['company', 'year'])[["gross"]].sum()

# Reset index to make 'company' and 'year' regular columns
CompanyGrossSumYear = CompanyGrossSumYear.reset_index()

# Sort values as required
CompanyGrossSumYearSorted = CompanyGrossSumYear.sort_values(['gross', 'company', 'year'], ascending=False)[:15]

# Convert 'gross' to integer if needed
CompanyGrossSumYearSorted['gross'] = CompanyGrossSumYearSorted['gross'].astype('int64')

CompanyGrossSumYearSorted
```
